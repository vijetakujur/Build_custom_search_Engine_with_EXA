```python
!pip install exa_py

from exa_py import Exa

exa = Exa('a5cd49e6-1b3f-4c72-9aa4-9ac3e4dad3b4')



# Getting started with Searching:

query = input('Search here: ')



response = exa.search(

  'Best Chicago cold brew',

  num_results=10,

)



print(response)



response2 = exa.search(

  'best pizza in Brooklyn',

  num_results=10,

  start_published_date='2023-05-01',

  category='tweet',

  use_autoprompt=True,

)



print(response2)



response3 = exa.search(

    query,

    num_results=5,

    type='keyword',

    include_domains=['https://www.tiktok.com']

)

print(response3)



for result in response3.results:

  print(f'Title: {result.title}')

  #print(f'Description: {result.description}')

  print(f'URL: {result.url}')

  print()
```
